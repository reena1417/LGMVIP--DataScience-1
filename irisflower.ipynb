```python
#Importing all the essintial libraries for our project

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
```

```python
#READIND THE DATA FROM THE GIVEN DATASET
```

```python
columns = ['sepal length','sepal width','petal length','petal width','class_label']
#load dataset
df = pd.read_csv('iris.data', names=columns)
df.head()
```

```python
df.describe()
```

```python
#VISUALIZATION OF DATA
```

```python
sns.pairplot(df, hue ='class_label')
```

```python
#SEPERATING INPUT COLUMN AND OUTPUT COLUMN
```

```python
#Seperate features and target

data = df.values

X = data[:,0:4]
y = data[:,4]

print(y)
```

```python
#SPLITING THE DATA INTO TRANING AND TESTING
```

```python
#Split the data into train and test

from sklearn.model_selection import train_test_split
X_train, x_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
print(x_test)
```

```python
#MODEL1: SUPPORT VECTOR MACHINE
```

```python
#Support vector machine algorithm
from sklearn.svm import SVC

model_svc = SVC()
model_svc.fit(X_train,y_train)
```

```python
prediction1 = model_svc.predict(x_test)
#calculate the accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,prediction1)*100)
for i in range(len(prediction1)):
  print(y_train[i],prediction1[i])
```

```python
#MODEL2: LOGISTIC REGRESSION ALGORITHM
```

```python
#Logistic regression algorithm

from sklearn.linear_model import LogisticRegression

model_LR = LogisticRegression()
model_LR.fit(X_train,y_train)
```

```python
prediction2 = model_LR.predict(x_test)
#calculate the accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,prediction2)*100)
for i in range(len(prediction2)):
  print(y_train[i],prediction2[i])
```

```python
#A DETAIL CLASSIFICATION REPORT
```

```python
# A detail classfication report
from sklearn.metrics import classification_report
print(classification_report(y_test , prediction2))
```
